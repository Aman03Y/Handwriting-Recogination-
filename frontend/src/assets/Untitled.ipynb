```python
# backend.py
import boto3

def recognize_handwriting(image_bytes):
    client = boto3.client('textract', region_name='ap-south-1')
    
    response = client.detect_document_text(
        Document={
            'Bytes': image_bytes
        }
    )
    
    recognized_text = ''
    for item in response['Blocks']:
        if item['BlockType'] == 'LINE':
            recognized_text += item['Text'] + '\n'
    
    return recognized_text

if __name__ == '__main__':
    # Example usage
    with open('photo3.png', 'rb') as file:
        image_bytes = file.read()
    
    text = recognize_handwriting(image_bytes)
    print(text)
```
